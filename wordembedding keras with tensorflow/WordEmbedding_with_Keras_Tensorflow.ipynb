```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences

sentence=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
# Define the vocabulary size

vocabulary_size=10000
```

```python
## one hot representation

one_hot_repr=[one_hot(i, vocabulary_size)for i in sentence]
one_hot_repr
```

```python
# Wordd Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# pad sequences used for suppose if we have 5 words and we convert it into 8 that use padding to make equal through
# pre=means before and post means after

sent_length=8  # we custom max length because in our dataset it have 5 max lenwords
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
### feature representation

dim=10
```

```python
model=Sequential()
model.add(Embedding(vocabulary_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
